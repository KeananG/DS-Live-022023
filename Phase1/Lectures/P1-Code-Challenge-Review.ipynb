```python
import pandas as pd
import requests
import ipywidgets as widgets
import pickle

def navigate_data(data):
    if isinstance(data, str):
        if data.endswith('.p'):
            with open(data, 'rb') as f:
                data = pickle.load(f)
        elif data.endswith('.csv'):
            data = pd.read_csv(data)
        else:
            response = requests.get(data)
            data = response.json()

    def display_key_value(key, value):
        output.clear_output()
        if isinstance(value, (dict, list)):
            keys = list(value.keys()) if isinstance(value, dict) else list(range(len(value)))
            dropdown.options = keys
            with output:
                display(dropdown)
        else:
            with output:
                display(value)

    def update_dropdown(change):
        key = change.new
        value = data[key]
        display_key_value(key, value)

    dropdown = widgets.Dropdown(options=[], description='Select key:')
    dropdown.observe(update_dropdown, names='value')
    output = widgets.Output()
    display(dropdown)
    display(output)
```
